```python
import pandas as pd
import numpy as np
```

```python
dpath = '/content/drive/MyDrive/Seculayer/OSC/data/for_ensemble/'
```

```python
ens1 = pd.read_csv(dpath+'KD_with_opt_submission.csv')
ens2 = pd.read_csv(dpath+'KD_with_core_opt_submission.csv')
ens3 = pd.read_csv(dpath+'KD_with_core_opt_version2_submission.csv')
```

```python
pred = (np.array(ens1['Y_LABEL']) + np.array(ens2['Y_LABEL']) + np.array(ens3['Y_LABEL'])) / 3
pred
```

```python
submission = pd.read_csv('/content/drive/MyDrive/Seculayer/OSC/data/sample_submission.csv')
```

```python
from collections import Counter
answer = np.zeros(pred.shape[0])

answer = np.where(pred>=0.12,1,0)
    
answer = answer.astype('int64')
print(Counter(answer))
```

```python
def make_submission(threshold):
    answer = np.zeros(pred.shape[0])
    answer = np.where(pred>=threshold,1,0)
    answer = answer.astype('int64')

    submission['Y_LABEL'] = answer
    submission.to_csv(f'/content/drive/MyDrive/Seculayer/OSC/data/submission/ensemble_{threshold}.csv',index=False)
```

```python
make_submission(0.1)
make_submission(0.11)
make_submission(0.12)
make_submission(0.13)
```

```python
make_submission(0.15)
```
