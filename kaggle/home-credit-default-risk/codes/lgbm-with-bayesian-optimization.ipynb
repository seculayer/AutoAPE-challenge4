```python
from pycaret.classification import *

clf = setup(app_train, 
            target = 'y_label', 
            categorical_features = [],
            numeric_features = numeric_list,
            normalize = False,
            iterative_imputation_iters = 10,
            train_size = 0.8,
            session_id = 42)
```

```python
best5 = compare_models(fold = 5, sort = 'AUC', n_select = 5)
```

```python
train = app_train.drop(columns = ['TARGET'])
train_labels = app_train['TARGET']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, train_labels, test_size=0.2)
```

```python
def objective(trial : Trial) -> float :

    params_cat = {
        'learning_rate' : trial.suggest_loguniform('learning_rate', 0.001, 1),
        'colsample_bytree' : trial.suggest_uniform('colsample_bytree',0,1),
        "n_estimators" : trial.suggest_int("n_estimators", 1000, 100000),
        "num_leaves" : trial.suggest_int("num_leaves", 10, 50),
        'subsample' : trial.suggest_uniform('sub_sample',0,1),
        'max_depth' : trial.suggest_int("max_depth", 5, 20),
        'reg_lambda': trial.suggest_uniform('reg_lambda',0,0.3),
        'reg_alpha': trial.suggest_uniform('reg_alpha',0,0.3),
        'min_split_gain': trial.suggest_uniform('min_split_gain',0,0.5),
        
  }
    
    model = LGBMClassifier(**params_cat,nthread=4,min_child_weight=39.3259775,silent=-1,verbose=-1)
    model.fit(X_train, y_train, eval_set = [(X_test, y_test)],eval_metric="auc",
              verbose = 100,early_stopping_rounds=200)

    cat_pred = model.predict(X_test)
    AUC = roc_auc_score(y_test, cat_pred)
    
    return AUC
```

```python
# Optuna 초모수 작업 시작

sampler = TPESampler(seed = 2023)
study = optuna.create_study(
    study_name = "LGBM_OPT",
    direction = "maximize",
    sampler = sampler)
study.optimize(objective, n_trials = 300)
```

```python
# 가장 좋은 초모수와 성능 확인

print("Best Score :", study.best_value)
print("Best trial :", study.best_trial.params)
```
